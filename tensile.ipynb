```python
import pandas as pd
import pathlib
import matplotlib.pyplot as plt
```

```python
data=[]
for name in pathlib.Path(".").glob("*.tbl"):
    data.append((name,pd.read_table(name, skiprows=4)))
```

```python
for name, tbl in data:
    plt.plot(tbl["dL(mm)"],tbl["F(Kn)"],label=name)
plt.legend()
```
